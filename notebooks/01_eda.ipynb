```python
# üìä Plot Null Value Distribution

# Null values per column
null_values = df.isna().sum()

fig, ax = plt.subplots(figsize=(18,7))

higher_nulls = null_values[null_values > 200].sort_values(ascending=False)

# Plot columns with significant missing values
higher_nulls.plot(kind='bar', ax=ax, color='teal')

ax.set_title('Missing Values Count', fontsize=18, fontweight='bold')
ax.set_xticklabels(higher_nulls.index, rotation=45)
ax.set_xlabel('Columns', fontsize=16, fontweight='semibold')
ax.set_ylabel('Count', fontsize=16, fontweight='semibold')
ax.set_ylim(0, 13500)
ax.set_yticks(range(0, 13001, 1000))
ax.grid(linewidth=0.2)

plt.show()
```

```python
# üìä Null Values Summary per Column

# Count null values per column
null_values = df.isna().sum()

# Total number of rows
total_rows = df.shape[0]

# Convert total rows into a Series for concatenation
total_rows_series = pd.Series(total_rows, index=df.columns)

# Calculate percentage of missing values
null_percentage = round(null_values / total_rows * 100, 2)

# Create summary DataFrame
df_summary = pd.concat([null_values, total_rows_series, null_percentage], 
                        keys=['Null Count', 'Total Rows', '% Null'], axis=1)

df_summary
```

```python
# üö® Columns with Excessive Missing Values (>70%)
incomplete_columns = df_summary[df_summary['% Null'] > 70]
incomplete_columns
```

```python
# üóëÔ∏è Drop Columns with Excessive Missing Values (>70%)
df = df.drop(incomplete_columns.index, axis=1)
```

## EDA (unitary, binary, multi)

### Tendencias globales:

Evoluci√≥n de low_carbon_electricity vs. fossil_electricity desde 1990 a la fecha.

```python
# Relation
# df[['low_carbon_electricity', 'fossil_electricity']]

low_carbon_electricity = df.groupby('year')['low_carbon_electricity'].sum()
fossil_electricity = df.groupby('year')['fossil_electricity'].sum()

plt.figure(figsize=(16,5))
plt.plot(df['year'].unique(), low_carbon_electricity, label='low carbon electricity')
plt.plot(df['year'].unique(), fossil_electricity, label='fossil electricity')
plt.legend()
plt.xlabel('Year', fontsize='16')
plt.ylabel('World Consume', fontsize='16')
plt.xticks(range(1950, 2030, 5))
plt.xlim(1950, 2030)
plt.grid(linewidth=0.3)
plt.show()
```

```python
# using seaborn
import seaborn as sns

top_10_2020 = renewables_share_elec_obs_2020.nlargest(15, 'renewables_share_elec')

plt.figure(figsize=(18, 6))
sns.set_theme(style='whitegrid')
ax = sns.barplot(data=top_10_2020, x='country', y='renewables_share_elec', hue='country', width=0.5, palette='viridis_r')
plt.xticks(rotation=45)

# 7. A√±adir etiquetas de datos sobre las barras (muy √∫til) (from Gemini 2.5 Pro)
for bar in ax.patches:
    # ax.patches es la colecci√≥n de todas las barras en el gr√°fico
    ax.annotate(f'{bar.get_height():.1f}%', # Texto a mostrar (altura de la barra, 1 decimal, con %)
                (bar.get_x() + bar.get_width() / 2., bar.get_height()), # Posici√≥n (x, y) del texto
                ha='center',          # Alineaci√≥n horizontal (centrada)
                va='center',          # Alineaci√≥n vertical (centrada)
                xytext=(0, 8),       # Desplazamiento en puntos (0 horizontal, 8 vertical para estar encima)
                textcoords='offset points',
                fontsize=9,           # Tama√±o de la fuente de la etiqueta
                color='black')  

plt.tight_layout()
plt.show()
```

### ¬øgdp est√° correlacionado con primary_energy_consumption? Usa gr√°ficos de dispersi√≥n.

```python
figure, ax = plt.subplots(1,2, figsize=(16,6))
sns.set_theme(style='whitegrid')
sns.scatterplot(data=df, ax=ax[0], x='gdp', y='primary_energy_consumption')
ax[0].set_title('GDP and PEC relation', fontsize=16, fontweight='bold')
ax[0].set_xlabel('Primary Energy Consuption')
ax[0].set_ylabel('GDP')

sns.scatterplot(data=df, ax=ax[1], x='fossil_energy_per_capita', y='greenhouse_gas_emissions')
ax[1].set_title('FEPC and GGE relation', fontsize=16, fontweight='bold')
ax[1].set_xlabel('Fossil energy per capita')
ax[1].set_ylabel('Greenhouse gas emissions')

plt.subplots_adjust(wspace=0.3)
plt.show()
```
